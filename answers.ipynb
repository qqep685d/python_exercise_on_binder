## FASTAファイル処理 - 各配列の情報をまとめる

FASTAファイル（[sample.txt](sample.txt)）には、20個分のシーケンスが入っています。 

それぞれのシーケンスの下記情報を取りまとめてください。  
- 配列名
- 配列の長さ
- アデニン「A」の個数
- チミン「T」の個数
- シトシン「C」の個数
- グアニン「G」の個数
- 未定義塩基「N」の個数

取りまとめたは情報は「タブ区切りファイル」に書き出してください。  
（出力ファイル名はなんでも構いません。）

![fileIO](fig/fileIO.png "fileIO")

```python
f_in = open('sample.txt', 'r') # FASTAファイル（入力ファイル）
f_out = open('fasta_summary.txt', 'w') # 出力ファイル

# 1行目かどうかのスイッチ（1=>1行目、0=>1行目以外）
first_line = 1

# 各塩基の個数
sumA = 0
sumT = 0
sumC = 0
sumG = 0
sumN = 0
# 配列の長さ
seqlen = 0 

# 一行ずつ読み込んで処理する
for line in f_in:
    line = line.rstrip()  # 改行コード削除
    
    # 配列名の行の時
    if '>' in line:
        
        # 一行目の時は何も情報を出力
        if first_line==1:
            first_line = 0 # スイッチを切り替える
            
        # 一行目以外の時、情報を出力する
        else:
            # 一つ前の配列の情報（蓄積した値）を出力
            print(seqlen, sumA, sumT, sumC, sumG)
            # ファイルに出力
            f_out.write("%s\t%s\t%s\t%s\t%s\n" % (seqlen, sumA, sumT, sumC, sumG))
            
            # 各変数の値を初期化
            sumA = 0
            sumT = 0
            sumC = 0
            sumG = 0
            sumN = 0
            seqlen = 0 
        
        # 配列名を表示
        print(line)
        # ファイルに出力
        f_out.write("%s\t" % line)

    # 配列名の行ではない（配列本体）の時
    else:
        # 一行に含まれる各塩基の個数をカウントして、それぞれの変数に入れる
        sumA = sumA + line.count("A")
        sumT = sumT + line.count("T")
        sumC = sumC + line.count("C")
        sumG = sumG + line.count("G")
        # 一行の塩基の合計数
        seqlen = seqlen + len(line)

# 最後の配列の情報を出力
print(seqlen, sumA, sumT, sumC, sumG)
f_out.write("%s\t%s\t%s\t%s\t%s\n" %  (seqlen, sumA, sumT, sumC, sumG))

# ファイルを閉じる
f_in.close()
f_out.close()
```
